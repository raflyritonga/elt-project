```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/data_engineering_project/imdb_movie_data_cleaning/imdb-movies-dataset.csv')
```

```python
df
```

```python
print(df.dtypes)
```

```python
# drop the unnecessary columns
df.drop(['Poster', 'Duration (min)', 'Metascore', 'Director', 'Votes', 'Description', 'Review Count', 'Review Title', 'Review' ], axis=1, inplace=True)
```

```python
# calculate the missing value of each row in each column
print(df.isnull().sum())
```

```python
# Dropping rows with null values in 'Year', 'Genre', 'Rating', and 'Cast' columns
df = df.dropna(subset=['Year', 'Genre', 'Rating', 'Cast'])
print("\nDataFrame after dropping rows with nulls in specific columns:\n", df)
```

```python
# Replacing null values in 'Certificate' column with 'none'
df['Certificate'] = df['Certificate'].fillna('none')
print("\nDataFrame after replacing nulls in 'Certificate' column:\n", df)
```

```python
df['Year'] = df['Year'].astype(int)
df
```

```python
df['movie_id'] = range(1, len(df) + 1)
df
```

```python
print(f"total rows: {df.shape[0]}")
print(df.isnull().sum())
```

```python
# Extracting the first cast member
df['First Cast Member'] = df['Cast'].apply(lambda x: x.split(', ')[0])
print("\nDataFrame with First Cast Member:\n", df)
```

```python
# Creating a new DataFrame with Movie ID and Cast Member, then renaming the columns
movie_and_cast_df = df[['movie_id', 'First Cast Member']].copy()
movie_and_cast_df.rename(columns={'First Cast Member': 'cast'}, inplace=True)
print("\nNew DataFrame with Movie ID and First Cast Member (renamed as cast_id):\n", movie_and_cast_df)
```

```python
# Creating a DataFrame of unique actors/actresses
unique_cast = df['First Cast Member'].unique()
actors_df = pd.DataFrame(unique_cast, columns=['actor_name'])
actors_df['id'] = actors_df.index + 1
print("\nActors DataFrame with unique actor names and IDs:\n", actors_df)
```

```python
# Merging movie_and_cast_df with actors_df to get the castfrom actors_df
movie_and_cast_df = movie_and_cast_df.merge(actors_df, left_on='cast', right_on='actor_name', how='left')
movie_and_cast_df = movie_and_cast_df[['movie_id', 'id']]
movie_and_cast_df.rename(columns={'id': 'actor_id'}, inplace=True)
print("\nFinal Movie and Cast DataFrame with movie_id and cast:\n", movie_and_cast_df)
```

```python
# Extracting the first genre
df['First Genre'] = df['Genre'].apply(lambda x: x.split(', ')[0])
print("\nDataFrame with First Genre:\n", df)
```

```python
# Creating a new DataFrame with Movie ID and Cast Member, then renaming the columns
movie_and_genre_df = df[['movie_id', 'First Genre']].copy()
movie_and_genre_df.rename(columns={'First Genre': 'genre_name'}, inplace=True)
print("\nNew DataFrame with movie id and its genre:\n", movie_and_genre_df)
```

```python
# Creating a DataFrame of unique actors/actresses
unique_genre = df['First Genre'].unique()
genre_df = pd.DataFrame(unique_genre, columns=['genre_name'])
genre_df['id'] = genre_df.index + 1
print("\nGenre DataFrame with genre and IDs:\n", genre_df)
```

```python
# Merging movie_and_genre_df with genre_df to get the genre_id from genre_df
movie_and_genre_df = movie_and_genre_df.merge(genre_df, left_on='genre_name', right_on='genre_name', how='left')
movie_and_genre_df = movie_and_genre_df[['movie_id', 'id']]
movie_and_genre_df.rename(columns={'id': 'genre_id'}, inplace=True)
print("\nFinal Movie and Cast DataFrame with movie_id and cast_id:\n", movie_and_genre_df)
```

```python
# drop the unused columns
df.drop(['Genre', 'Cast', 'First Cast Member', 'First Genre', 'movie_id' ], axis=1, inplace=True)
genre_df.drop('id', axis=1, inplace=True)
actors_df.drop('id', axis=1, inplace=True)
```

```python
# rename the columns
df.rename(columns={'Title': 'title', 'Year': 'released_year', 'Certificate': 'imdb_certificate', 'Rating': 'rating'}, inplace=True)
```

```python
# dataframes to CSVs
df.to_csv('imdb_movie.csv', index=False)
genre_df.to_csv('genre.csv', index=False)
actors_df.to_csv('actor.csv', index=False)
movie_and_cast_df.to_csv('movie_and_actor.csv', index=False)
movie_and_genre_df.to_csv('movie_and_genre.csv', index=False)
```
